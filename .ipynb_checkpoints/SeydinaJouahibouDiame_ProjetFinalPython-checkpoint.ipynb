```python
def pourcentage(ser):
    return (ser/float(ser[-1]))*100
```

##### Tableaux de contingence entre la variable "Do you celebrate Thanksgiving" qui est la variable dépendante et la variable "How much total combined money did all members of your HOUSEHOLD earn last year?"

###### Ma reponse

```python
FamillyMoney_Thanksgiving = pd.crosstab(data['Do you celebrate Thanksgiving?']  , 
                            data['How much total combined money did all members of your HOUSEHOLD earn last year?'],
                                margins = True).apply(pourcentage, axis=1)

FamillyMoney_Thanksgiving
```

Pour savoir si deux variables catégorielles sont liées , on utilise le célébre test du khi-deux

```python
import pandas as pd 
from scipy.stats import chi2_contingency

from scipy.stats import chi2_contingency as chi2_contingency

resultats_test1 = chi2_contingency(FamillyMoney_Thanksgiving)
print("Statistique de test : ",resultats_test1[0])
print("p_valeur : ",resultats_test1[1])
print("degré de liberté : ",resultats_test1[2])
```

Dans notre exemple ci-dessus la p-valeur est bien inférieur à 5% on peut donc rejeter l’hypothèse selon laquelle les deux variables testées sont indépendantes.

## III.4 Convertir l'âge en une valeur numérique

- Ecrire une fonction qui convertit une chaine de caractères en une valeur entière. Cela permettra de convertir les valeurs de la colonne 'Age' en entiers. Cette fonction prendra en paramètre une chaine de caractères (les valeurs actuelles de la colonne 'Age')
>- Utiliser la fonction isnull() pour vérifier si les valeurs sont nulles. Ajouter une condition if qui retourne 'None' si la valeur est nulle
>- Séparer les chaines de caractères en fonction de l'espace (' ') et extraire le premier élément de la liste résultante (méthode split())
>- Remplacer le caractère '+' dans le résultat avec une chaine de caractères vide en remplacement pour le supprimer (méthode replace())
>- utiliser int() pour convertir le résultat en entier
>- Retourner le résultat
- Utiliser la méthode Series.apply() pour appliquer la fonction à chaque valeur de la colonne 'Age' du dataframe data.
>- Assigner le résultat à la nouvelle colonne int_age du dataframe data
- Appeler la méthode Series.describe() sur la colonne 'int_age' du dataframe data et afficher le résultat.


###### Ma reponse

```python

def age(colonne):
    donne = data[colonne]
    for i in range(len(donne)):
        if pd.isnull(donne[i]):
            donne[i] = None
        else :
            donne[i] = donne[i].split()[0].replace("+","")
            data['int_age'][i] =(float(donne[i]))
                        
age("Age")
data["int_age"].describe()



      
```

## III.5 Convertir les revenus en valeurs numériques
